#### Question 3a (1.5 marks)

Complete the code to read in names and accounts from the file accounts.txt. **(1/2 mark per field)**

```python
import re

addressRegex = re.compile(r'''(
            (\w+)
            (\w{2})             
            .\s
            (.\w+)
            .\s
            (.\d+)
            .\s
            (.\d+.\d+.\d+)
            .\s
            (.\d+)
            .\s
            (.\d+.\d+.\d+)
            )''', re.VERBOSE)     # update the preceding lines for regex (2 marks)

with open("accounts.txt") as fileobj:
    data = fileobj.readlines()

names = []   # use regular expressions to extract this separately
bank = []    # first three digits of BSB - account number
branch = []  # second three digits of BSB 
account = [] # specific account number

for address in data:
        address = address.strip()
        mo = addressRegex.search(address)
        if mo:
            print('Matched: ' + mo.group())
            names.append(mo.group(2))
            bank.append(mo.group(3))
            branch.append(mo.group(4))
            account.append(mo.group(5))     # extract data like this

print("\nName\t\tBank\tBranch\tAccount") # Headings 
print("=================================================") 

# print the data
for i in range(len(names)):
    print(names[i] + "\t" + bank[i] + "\t" + branch[i] + "\t" + account[i])
    
            
```

#### Question 4c (4 marks)

The following `growthplot.py` code uses command line arguments for simulation `length`, `population` and `growthrate` to allow it to be run in a parameter sweep (these are commented out to run it in Jupyter Notebook)

Modify the code to save the output and avoid interaction during a parameter sweep.
1. add titles and labels, then save the plot to a file, named to match the parameters **(1 mark)**
2. save the printed information to a **valid** csv file, named to match the parameters **(2 marks)**
3. suppress any other output **(1 mark)**

```python
'''
growthplot.py - simulation of unconstrained growth
'''
import numpy as np
import matplotlib.pyplot as plt
import csv

print("\nSIMULATION - Unconstrained Growth\n")
length = 20           # int(sys.argv[1]) - commented out for jupyter notebook
population = 100      # int(sys.argv[2])
growth_rate = 0.1     # float(sys.argv[3])
time_step = 1
num_iter = int(length / time_step)
growth_step = growth_rate * time_step

timearray = np.zeros(num_iter +1)
poparray = np.zeros(num_iter +1)

timearray[0] = 0
poparray[0] = population

print("Time: ", 0, " \tGrowth: ", 0, " \tPopulation: ", 100)

for i in range(1, int(num_iter) + 1 ):
    growth = growth_step * population
    population = population + growth
    i_time = i * time_step
    timearray[i] = i_time
    poparray[i] = population
    print("Time: ", i_time, " \tGrowth: ", growth," \tPopulation: ", population)
    #write outputs time, growth, population to csv file
    with open('Unconstrained_Growth_Outputs.csv', 'a') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Time', i_time, 'Growth', growth, 'Population', population])

print("\nPROCESSING COMPLETE.\n")

plt.title("Unconstrained Growth")
plt.plot(timearray, poparray)
plt.xlabel("Time")
plt.ylabel("Population")
plt.savefig('Unconstrained_Growth.png')
plt.show()
```
